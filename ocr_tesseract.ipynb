```
def preprocess(img):
    ty = ['.png','.png','.jpg']
    test_images = []
    for i,image in enumerate(img):
        preprocessed_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        preprocessed_img = cv2.threshold(preprocessed_img ,0,255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]
        preprocessed_img = cv2.medianBlur(preprocessed_img , 3)
        s = 'temp_img' + str(i) + ty[i]
        #print(s)
        cv2.imwrite(s,preprocessed_img)
        preprocessed_img_pil_img = Image.open(s)
        test_images.append(preprocessed_img_pil_img)
    
    return test_images


'''
preprocessed_img = cv2.cvtColor(image_to_ocr, cv2.COLOR_BGR2GRAY)
preprocessed_img = cv2.threshold(preprocessed_img,0,255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]
preprocessed_img = cv2.medianBlur(preprocessed_img, 3)
cv2.imwrite('temp_img.jpg',preprocessed_img)
preprocessed_img_pil_img = Image.open('temp_img.jpg')
text_extracted = pytesseract.image_to_string(preprocessed_img_pil_img)
print(text_extracted)'''
test_images_arr = preprocess(image)

for i in range(3):
    text_extracted = pytesseract.image_to_string(test_images_arr[i])
    print(text_extracted)

    plt.imshow(image[i])
    plt.show()
```
